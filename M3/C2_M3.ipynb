**Estilo de linea**  

Las opciones disponibles son:

"-" o "solid": l√≠nea s√≥lida (es la opci√≥n por defecto)  
"--" o "dashed": l√≠nea discontinua (--)  
"-." o "dashdot": l√≠nea que alterna guiones y puntos  
":" o "dotted": l√≠nea de puntos  
"None", " " o "": no muestra nada

```python
plt.plot(x,linestyle=":")
plt.show()
```

```python
#equivalencia
fig = plt.figure()   #definimos nuestra figura
axes=plt.axes()
axes.plot(x,linestyle=":") #agregamos elementos al contenido
plt.show()  
```

```python
#

fig = plt.figure(edgecolor="blue",linewidth=15)
fig.set_size_inches(20,5)
plt.plot(np.random.rand(50))
plt.suptitle("recorrido aleatorio",fontsize=20)
plt.show()
```

Una forma de ponerle t√≠tulo a toda la figura ( No gr√°fico por gr√°fico es con el m√©todo suptitle de figure)  

```python
#
fig, ax= plt.subplots(1,2)
fig.set_size_inches(20,5)
fig.suptitle("recorrido aleatorio",fontsize=20)
ax[0].plot(np.random.rand(50),color="red")
ax[1].plot(np.random.rand(50))
```

Otros par√°metros importantes son el tama√±o de nuestro t√≠tulo ( ojo, aqu√≠ uso plt.title y no fig.suptitle porque quiero el t√≠tulo del gr√°fico y no de la figura), su tama√±o, su alineaci√≥n, as√≠ como su posici√≥n en la figura


```python

fig, ax= plt.subplots()
fig.set_size_inches(20,5)
plt.title("recorrido aleatorio",fontsize=20,horizontalalignment="left",position=(.5,.5),color="purple",fontweight="bold")
ax.plot(np.random.rand(50),color="red")
```

#### T√≠tulos en Ejes

Para poder asignar los t√≠tulos a los ejes necesitamos hacerlo por grafico ( no por figura ) con lo que tenemos que llamar al m√©todo xlabel del gr√°fico (plt).

Al igual que los dem√°s m√©todos, comparten ciertos par√°metros como son tama√±o de letra y color, adem√°s de que podemos asignar color a nuestro contorno


```python

fig, ax= plt.subplots()
fig.set_size_inches(20,5)
plt.title("recorrido aleatorio",fontsize=20,horizontalalignment="left",position=(.5,.5),color="purple",fontweight="bold")
plt.xlabel("eje1",fontsize=20,color="white",backgroundcolor="black")
plt.ylabel("eje2",fontsize=20,color="white",backgroundcolor="black")
ax.plot(np.random.rand(50),color="red")
```

#### Alcance de los ejes

#### Etiquetas de datos
Finalmente para poder asignar las etiquetas de nuestros datos tenemos que asignar en nuestro axes el t√≠tulo de nuestra etiqueta as√≠ como ejecutar el m√©todo legend de axes donde alinearemos las mismas


```python
 
fig,ax=plt.subplots()
ax.plot(np.random.rand(20).cumsum(),linestyle="-",linewidth=2,marker="o",markersize=1 ,color="purple",label="Gr√°fico 1")
ax.plot(np.random.rand(20).cumsum(),linestyle="-",linewidth=2,marker="v",markersize=1 ,color="red",label="Gr√°fico 2")
ax.plot(np.random.rand(20).cumsum(),linestyle="-",linewidth=2,marker="3",markersize=1 ,color="green",label="Gr√°fico 3")
ax.legend(["Gr√°fico 1","Gr√°fico 2","Gr√°fico 3"],loc=(0.2,0.6))

plt.show()
```

Con el par√°metro n col definimos en cuantas columnas queremos que se desplieguen nuestras etiquetas

```python


fig,ax=plt.subplots()
ax.plot(np.random.rand(20).cumsum(),linestyle="-",linewidth=2,marker="o",markersize=1 ,color="purple",label="Gr√°fico 1")
ax.plot(np.random.rand(20).cumsum(),linestyle="-",linewidth=2,marker="v",markersize=1 ,color="red",label="Gr√°fico 2")
ax.plot(np.random.rand(20).cumsum(),linestyle="-",linewidth=2,marker="3",markersize=1 ,color="green",label="Gr√°fico 3")
ax.legend(["Gr√°fico 1","Gr√°fico 2","Gr√°fico 3"],loc=(0.2,0.6),ncol=3)

plt.show()
```

Finalmente vemos que el estilo de las etiquetas se puede definir totalmente en el metodo `legends` de axes, aqui podemos asignarle a nuestras etiquetas columnas, tama√±o de letra, un titulo al marco de etiquetas, su locacion etc.
